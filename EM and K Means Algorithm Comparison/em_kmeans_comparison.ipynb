```python
import matplotlib.pyplot as plot
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn import preprocessing
from sklearn.mixture import GaussianMixture
import pandas as pd
import numpy as np
```

```python
# Iris is a flowering plant.
iris = datasets.load_iris()
```

```python
data = pd.DataFrame(iris.data)
data.columns = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width']

print(data)
```

```python
targets = pd.DataFrame(iris.target)
targets.columns = ['Targets']

print(targets)
```

```python
# for coloring dots / clusters
colorMap = np.array(['red', 'green', 'blue', 'yellow'])
```

```python

# REAL CLUSTERS
plot.scatter(data.Petal_Length, data.Petal_Width, c = colorMap[targets.Targets], s=5) # s - size of dot

plot.title('Real Clusters')
plot.xlabel('Petal Length')
plot.ylabel('Petal Width')
```

```python

# K MEANS CLUSTERING
model = KMeans(n_clusters = 4)
model.fit(data)

plot.scatter(data.Petal_Length, data.Petal_Width, c = colorMap[model.labels_], s=5)
plot.title('K Means Clustering')
plot.xlabel('Petal Length')
plot.ylabel('Petal Width')
```

```python

# Gaussian Mixture Model CLUSTERING
scaler = preprocessing.StandardScaler()
scaledData = pd.DataFrame(scaler.fit(data).transform(data), columns = data.columns)

gmm = GaussianMixture(n_components = 4)
gmm.fit(scaledData)
gmm_targets = gmm.predict(scaledData)

plot.scatter(data.Petal_Length, data.Petal_Width, c = colorMap[gmm_targets], s=5)
plot.title('GMM Clustering')
plot.xlabel('Petal Length')
plot.ylabel('Petal Width')
```

```python
print("Observation:The GMM Clustering using EM Algorithm matched the True Labels more closely than the K Means Clustering")
```
